```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"  # Use a valid dataset URL
df = pd.read_csv(url)

# Ensure proper column names
df.columns = df.columns.str.strip()  # Remove any unexpected spaces in column names

# Fill missing Age values with the mean (Fixed Version)
df.loc[:, 'Age'] = df['Age'].fillna(df['Age'].mean())

# Filter survivors (Survived == 1)
survivors = df[df['Survived'] == 1]

# Scatter plot: Male vs Female survivors (Age vs Fare)
plt.figure(figsize=(10, 6))

# Male Survivors
plt.scatter(
    survivors[survivors['Sex'] == 'male']['Age'],
    survivors[survivors['Sex'] == 'male']['Fare'],
    color='blue', label='Male Survivors', alpha=0.6
)

# Female Survivors
plt.scatter(
    survivors[survivors['Sex'] == 'female']['Age'],
    survivors[survivors['Sex'] == 'female']['Fare'],
    color='red', label='Female Survivors', alpha=0.6
)

# Labels and title
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Scatter Plot of Male and Female Survivors (Age vs Fare)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```
