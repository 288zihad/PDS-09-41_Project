```python
import pandas as pd
import numpy as np
df= pd.read_csv('https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv')
df
# Checking columns with missing values
null_columns = df.columns[df.isnull().any()]
print("Columns with Null values:", null_columns)
```

```python
# Fill 'Age' column with the mean value
# Fill 'Cabin' column with 'Unknown'
df.fillna({'Age': df['Age'].mean(), 'Cabin': 'Unknown'}, inplace=True)
df
```

```python
# Bar plot: Survived vs Dead (Grouped by Gender)
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(data=df, x='Survived', hue='Sex', palette=['red', 'blue'])
plt.title('Survival Count by Gender')
plt.xticks([0, 1], ['Dead', 'Survived'])
plt.show()

# Bar plot: Survived vs Dead (Grouped by Pclass)
sns.countplot(data=df, x='Survived', hue='Pclass', palette='coolwarm')
plt.title('Survival Count by Passenger Class')
plt.xticks([0, 1], ['Dead', 'Survived'])
plt.show()
```

```python
# Create AgeClass column
df['AgeClass'] = pd.cut(df['Age'], bins=[0, 16, 26, 36, 62, np.inf], labels=[0, 1, 2, 3, 4])

# Drop the Age column
df.drop(columns=['Age'], inplace=True)
```

```python
sns.countplot(data=df, x='Survived', hue='AgeClass', palette='Set2')
plt.title('Survival Count by Age Class')
plt.xticks([0, 1], ['Dead', 'Survived'])
plt.show()
```
