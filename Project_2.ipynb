```python
# Creating term-test-1-result.csv
students = {
    "Registration Number": range(1001, 1051),
    "Name": [f"Student_{i}" for i in range(1, 51)],
    "TT-1 Marks": np.random.randint(5, 20, 50)
}
df1 = pd.DataFrame(students)
df1.to_csv("term-test-1-result.csv", index=False)
df1
```

```python
# Creating term-test-2-result.csv
students["TT-2 Marks"] = np.random.randint(5, 20, 50)
df2 = pd.DataFrame(students)
df2.to_csv("term-test-2-result.csv", index=False)
df2
```

```python
# Load CSV files
df1 = pd.read_csv("term-test-1-result.csv")
df2 = pd.read_csv("term-test-2-result.csv")

# Merge both dataframes
merged_df = pd.merge(df1, df2[['Registration Number', 'TT-2 Marks']], on='Registration Number')
merged_df
```

```python
# Calculate best and average term test marks
merged_df["Best Marks"] = merged_df[['TT-1 Marks', 'TT-2 Marks']].max(axis=1)
merged_df["Average Marks"] = merged_df[['TT-1 Marks', 'TT-2 Marks']].mean(axis=1)
merged_df
```

```python
merged_df.drop(columns=['TT-1 Marks', 'TT-2 Marks'], inplace=True)
```

```python
# Creating attendance and term final marks data
attendance_marks = np.random.randint(5, 10, 50)
term_final_marks = np.random.randint(40, 100, 50)

attendance_df = pd.DataFrame({
    "Registration Number": range(1001, 1051),
    "Attendance Marks": attendance_marks,
    "Term Final Marks": term_final_marks
})

# Merging with term test dataframe
final_df = pd.merge(merged_df, attendance_df, on="Registration Number")

# Calculate Final Marks
final_df["Final Marks"] = final_df["Term Final Marks"] * 0.7 + final_df["Average Marks"] + final_df["Attendance Marks"]
```

```python
final_df.to_csv("final_result.csv", index=False)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Select Final Marks for clustering
X = final_df[["Final Marks"]]

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
final_df["Cluster"] = kmeans.fit_predict(X)

# Visualization
plt.scatter(final_df["Registration Number"], final_df["Final Marks"], c=final_df["Cluster"], cmap='viridis')
plt.xlabel("Registration Number")
plt.ylabel("Final Marks")
plt.title("K-Means Clustering of Students' Final Marks")
plt.colorbar(label="Cluster")
plt.show()
```
